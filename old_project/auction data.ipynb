```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
t08= pd.read_html('https://en.wikipedia.org/wiki/List_of_2008_Indian_Premier_League_auctions_and_personnel_signings')
t11= pd.read_html('https://en.wikipedia.org/wiki/List_of_2011_Indian_Premier_League_personnel_changes')
t12=pd.read_html('https://en.wikipedia.org/wiki/List_of_2012_Indian_Premier_League_personnel_changes')
```

```python
final=[t08[2],t11[2],t12[6]]
f1=pd.concat(final, sort=True)
```

```python
f1.drop(['No.'],axis=1,inplace=True)
f1.head()
```
