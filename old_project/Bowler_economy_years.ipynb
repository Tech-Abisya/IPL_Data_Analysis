```python
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
```

```python
bowl_eco = pd.read_csv('yashfile.csv')
bowl_eco.drop(columns='Unnamed: 0',inplace=True)
bowl_eco
```

```python
bowl_sruns=bowl_eco.groupby(['bowler','season'])['total_runs'].sum()
bowl_sruns=pd.DataFrame(bowl_sruns)
#bowl_sruns.reset_index(inplace=True)
```

```python
bowl_sovers=bowl_eco.groupby(['bowler','season'])['over'].count()
bowl_sovers=pd.DataFrame(bowl_sovers)
#bowl_sovers.reset_index(inplace=True)
```

```python
bowl_s=pd.merge(bowl_sruns,bowl_sovers,on=['bowler','season'])
a = bowl_s['total_runs']/bowl_s['over']
avg=pd.DataFrame(a,columns=['avg_economy'])
bowler_avg=pd.concat([bowl_s,avg],axis=1)
bowler_avg.head()
```

```python
bowlers = bowler_avg.groupby(['bowler','season','total_runs','avg_economy'])
```

```python
bowlers=bowlers.agg('mean')
```

```python
ba=bowler_avg.pivot_table('avg_economy','bowler','season')
ba['av.eco']=ba.mean(axis=1,skipna=True)
bowler_avg=ba.sort_values(by='av.eco')
```

```python
tot_overs=bowl_eco.groupby('bowler')['over'].count()
tot_overs=pd.DataFrame(tot_overs)
tot_overs.reset_index(inplace=True)
tot_overs
```

```python
bowler__avg=pd.merge(bowler_avg,tot_overs,on='bowler')
```

```python
bowler__avg.to_csv('bavg.csv')
```
