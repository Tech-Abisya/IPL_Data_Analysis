```python
#bowlers average economy 
bowl=data.groupby(['match_id','bowler','over'])['total_runs'].sum()
bowl_eco=pd.DataFrame(bowl)

bowl_eco
```

```python
bowl_eco=bowl_eco.reset_index()
tot_overs=bowl_eco.groupby('bowler')['over'].count()
tot_overs=pd.DataFrame(tot_overs)
tot_overs=tot_overs.reset_index()
```

```python
tot_runs=bowl_eco.groupby('bowler')['total_runs'].sum()
tot_runs=pd.DataFrame(tot_runs)
tot_runs=tot_runs.reset_index()
tot_runs
```

```python
total_bowl=pd.merge(tot_overs,tot_runs,on='bowler')
#total_bowl.sort_values()
a = total_bowl['total_runs']/total_bowl['over']
avg=pd.DataFrame(a,columns=['avg_economy'])
bowler_avg=pd.concat([total_bowl,avg],axis=1)

#bowler_avg.sort_values('total_runs',ascending=False)
bowler_avg_backup=bowler_avg
bowler_avg.sort_values('avg_economy')
bowler_avg.to_csv('bowleravg1819.csv',index=False)
```
