```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
t08= pd.read_html('https://en.wikipedia.org/wiki/List_of_2008_Indian_Premier_League_auctions_and_personnel_signings')
t11= pd.read_html('https://en.wikipedia.org/wiki/List_of_2011_Indian_Premier_League_personnel_changes')
t14= pd.read_html('https://en.wikipedia.org/wiki/List_of_2014_Indian_Premier_League_personnel_changes')
t17= pd.read_html('https://en.wikipedia.org/wiki/List_of_2017_Indian_Premier_League_personnel_changes')
```

```python
s17_1 = t17[1].rename(columns={'Team':'Team 2017'})
s17_2 = s17_1.drop('Nat',axis=1)
s17_2.rename(columns={'Base price(in ₹ lakhs)':'Base.P 2017','Winning bid(in ₹ lakhs)':'Auction.P 2017'},inplace=True)
s17_2.head()
```

```python
s11_1 = t11[2].rename(columns={'Team':'Team 2011'})
s11_2 = s11_1.drop('Nat',axis=1)
columnt=['Player','Team 2011','Base.P 2011','Auction.P 2011']
s11_3=s11_2.reindex(columns=columnt)
s11_3.rename(columns={'Base price':'Base.P 2011','Winning bid':'Auction.P 2011'},inplace=True)
s11_3.head()
```

```python
s14_1=t14[6].drop(['Set No','Set','Country','IPL Matches','IPL Team(s)','Capped / Uncapped','IPL 2013 Team'],axis=1)
s14_1.rename(columns={'Reserve Price in Lakhs in INR':'Base.P 2014','Auctioned Price in Lakhs in INR':'Auction.P 2014','IPL 2014 Team':'Team 2014','Name':'Player'},inplace=True)
columnt=['Player','Team 2014','Base.P 2014','Auction.P 2014']
s14_2=s14_1.reindex(columns=columnt)
s14_2.head()
s14_2.to_csv('14.csv',index=False)
s14_3=pd.read_csv('14f.csv')
```

```python
s08_1=t08[2].drop(['No.','Category'],axis=1)
s08_1.rename(columns={'Base Price(in US$ thousands)':'Base.P 2008','Auctioned Price(in US$ thousands)':'Auction.P 2008','Team':'Team 2008','Name':'Player'},inplace=True)
columnt=['Player','Team 2008','Base.P 2008','Auction.P 2008','Nationality','Speciality']
s08_2=s08_1.reindex(columns=columnt)
s08_2.to_csv('8.csv',index=False)
s08_3=pd.read_csv('8f.csv')
```

```python
f1=pd.merge(s17_2,s11_3,on='Player',how='outer')
f2=pd.merge(f1,s14_3,on='Player',how='outer')
f3=pd.merge(f2,s08_3,on='Player',how='outer')
f3.head()
```

```python
f3.to_csv('ishita.csv',index=False)
```

```python
a1=pd.read_csv('ishita.csv')
b1=pd.read_csv('yash.csv')
c1=pd.read_csv('abhinav.csv')
```

```python
f11=pd.merge(a1,b1,on='Player',how='outer')
f2=pd.merge(f11,c1,on='Player',how='outer')
f2.tail()
```
