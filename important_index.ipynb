```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing 
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
s819=pd.read_excel('bowler/BowlerOuter.xlsx',sheet_name='819')
s1819=pd.read_excel('bowler/BowlerOuter.xlsx',sheet_name='1819')
s1617=pd.read_excel('bowler/BowlerOuter.xlsx',sheet_name='1617')
s1617.head()
```

```python
a1=s1617.drop(columns='bowler')
x = a1.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled,columns=['total_overs','total_runs','avg_economy','death_over','over','wide_runs','WoPerOver','noball_runs','NoPerOver','powerplay_over','wickets'])
df.head()
```

```python
df['value1617']=df['total_overs']*0.02 +df['avg_economy']*0.35 + df['wickets']*0.35 + df['WoPerOver']*0.1 +df['NoPerOver']*0.1 +df['death_over']*0.04 +df['powerplay_over']*0.04
```

```python
df.sort_values(by='value1617',ascending=False).head()
```

```python
s1617['values']=df['value1617']
s1617=s1617.sort_values(by='values',ascending=False)
s1617.to_csv('s1617ibi.csv',index=False)
```
