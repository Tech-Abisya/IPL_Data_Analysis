```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing 
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
s819=pd.read_excel('batsman/batsmanOuter.xlsx',sheet_name='819')
s1819=pd.read_excel('batsman/batsmanOuter.xlsx',sheet_name='1819')
s1617=pd.read_excel('batsman/batsmanOuter.xlsx',sheet_name='1617')
s819.head()
```

```python
a1=s1819.drop(columns='batsman')
x = a1.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled,columns=['avg.runs','4\'s','6\'s','50\'s','100\'s','no of balls','batsman_runs','strike','30\'s'])
df.head()
```

```python
df['value']=df['no of balls']*0.02 +df['batsman_runs']*0.10 + df['strike']*0.25 + df['avg.runs']*0.25 +df['50\'s']*0.05 +df['30\'s']*0.18 +df['4\'s'
]*0.075+df['6\'s']*0.075
```

```python
df.sort_values(by='value',ascending=False).head()
```

```python
s1819['values']=df['value']
s1819=s1819.sort_values(by='values',ascending=False)
s1819.to_csv('s1819ii.csv',index=False)
```
